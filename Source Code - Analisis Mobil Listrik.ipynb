# Modelling

Untuk memperkuat hasil analisis di atas, kami memanfaatkan model machine learning untuk memprediksi kondisi jumlah produksi dan penjualan BEV, angka oil displacement, dan rata-rata emisi CO2 yang dihasilkan per BEV dan motor car untuk satu dekade ke depan.

```python
# Data features
bev_columns = ['EV sales BEV', 'EV Stock BEV', 'Oil displacement (Million barrels per day)',
               'Oil displacement (Milion litres gasoline equivalent)']
X = df1[bev_columns]
y = df1['Average CO2 emission per BEV produced (Kg CO2e)']

columns_add = ['Total Motor cars production per year']
```

```python
# List model machine learning yang digunakan
from sklearn.linear_model import LinearRegression, Ridge, Lasso, ElasticNet
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from xgboost import XGBRegressor
from sklearn.svm import SVR
from sklearn.tree import DecisionTreeRegressor

models = {
    "Linear Regression": LinearRegression(),
    "Random Forest Regressor": RandomForestRegressor(),
    "Gradient Boosting Regressor": GradientBoostingRegressor(),
    "XGBoost Regressor": XGBRegressor()
}
```

```python
# Mencari model machine learning yang paling tepat untuk digunakan

from sklearn.model_selection import TimeSeriesSplit, cross_val_score

print("Below are MAE scores for each model:")

def run_cross_val(models, X, y):
  mae_models = list()
  for model in models:
    # Multiply by -1 since sklearn calculates *negative* MAE
    # Perform time series cross-validation
    tscv = TimeSeriesSplit(n_splits = 5)
    scores = -1 * cross_val_score(models[model], X, y,
                                  cv=tscv,
                                  scoring='neg_mean_absolute_error')

    print('----------------------------------')
    print(model)
    print("Average MAE score (across experiments):")
    mae = scores.mean()
    print(mae)
    print("Average RMSE score (across experiments):")
    print(np.sqrt(scores.mean()))
    mae_models.append(mae)
  return mae_models

mae_models_bev = run_cross_val(models, X, y)
```

```python
df_mae = pd.DataFrame({
    "Model": models.keys(),
    "MAE":  mae_models_bev
})
df_mae.sort_values("MAE").plot(x="Model", y="MAE",kind='bar')
plt.xticks(rotation=90)
plt.title("Perbandingan MAE Setiap Model")
plt.show()
```

Berdasarkan percobaan di atas, model Linear Regression menghasilkan akurasi paling baik sehingga prediksi-prediksi berikutnya akan menggunakan model tersebut.

```python
# Set linear regression sebagai chosen model
chosen_model = models["Linear Regression"]
```

```python
df = pd.DataFrame(df1)

# Membuat DataFrame baru untuk menyimpan predicted values
start_year = 2022
end_year = start_year + 10
next_decade_years = pd.Series([i for i in range(start_year, end_year+1)])
predicted_df = pd.DataFrame(index=pd.RangeIndex(start_year, end_year + 1, name='Year'))

for column in bev_columns + columns_add:
    # Get the X and y values for the linear regression
    X_new = df['Year'].values.reshape(-1, 1)
    y_new = df[column].values.reshape(-1, 1)

    # Create and fit the model
    model = LinearRegression()
    model.fit(X_new, y_new)

    next_decade_predictions = model.predict(next_decade_years.values.reshape(-1, 1))

    # Menambahkan predicted values ke DataFrame baru
    predicted_df[column] = next_decade_predictions.flatten()

predicted_df
```
